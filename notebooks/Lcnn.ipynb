```python
import numpy as np

# the conventional data load functions in the code

# path for dataset to load the dataset
dataset_path = 'dataset/'

""" 
The function takes the data path, and the text file name as input and returns the normalized matrix

Parameters:
	- dataset_path (str): The path to the dataset.
	- type_ (str): The type of dataset to load (e.g., 'train', 'test').
	- type (str): The type of data to load (e.g., 'shape', 'el', 'dist', 'hr').
Returns:
	- matrix (numpy.ndarray): The normalized matrix.
"""
# load a single file as a numpy array
def read_and_norm(dataset_path, type_, type):
	# read the txt files and save it as a numpy array
	with open('{}/{}/{}.txt'.format(dataset_path, type_, type), 'rb') as f:
		matrix = [[float(x) for x in line.split()] for line in f]
	matrix = np.array(matrix)
	
	# normalize the matrix
	min_m = matrix.min().min()
	max_m = matrix.max().max()
	matrix = ((matrix - min_m) / (max_m - min_m))
	return matrix

# load a dataset group, such as train or test
def load_full_dataset(type_):
	"""
	Load and preprocess the full dataset for a given type.

	This function loads the classification labels and the shape, el, dist, and hr data from the specified dataset path.
	It normalizes the shape, el, and dist data, removes rows with 9 columns from the classification and other data arrays,
	and combines the shape, dist, el, and hr data into a single dataset.

	Parameters:
	type_ (str): The type of dataset to load (e.g., 'train', 'test').

	Returns:
	tuple: A tuple containing:
		- data_X (numpy.ndarray): The combined dataset with shape, dist, el, and hr data.
		- classification (numpy.ndarray): The classification labels.
	"""

	# load the classification labels
	# create a numpy array from the classification file
	classification = np.loadtxt('{}/{}/classification.txt'.format(dataset_path, type_))
	classification = np.array(classification).reshape(-1, 1)

	# load the hydropathy data
	# create a numpy array from the hr file
	with open('{}/{}/hr.txt'.format(dataset_path, type_), "r") as file:
		hr = []
		righe_con_9_colonne = []
		for indice, riga in enumerate(file):
			colonne = riga.split()
			if len(colonne) == 9:
				righe_con_9_colonne.append(indice)
			else:
				hr.append(colonne)
	hr = [[float(string) for string in inner] for inner in hr]
	hr = np.array(hr)

	# read the shape, el, and dist data
	# normalize the shape, el, and dist data
	shape = read_and_norm(dataset_path, type_, 'shape')
	el = read_and_norm(dataset_path, type_, 'el')
	dist = read_and_norm(dataset_path, type_, 'dist')

	# remove rows with 9 columns from classification, shape, el, and dist
	classification = np.delete(classification, righe_con_9_colonne, 0)
	shape = np.delete(shape, righe_con_9_colonne, 0)
	el = np.delete(el, righe_con_9_colonne, 0)
	dist = np.delete(dist, righe_con_9_colonne, 0)

	# combine shape, dist, el, and hr into a single dataset
	data_X = np.array([p for p in zip(shape, dist, el, hr)])
	data_X = data_X.reshape(data_X.shape[0], data_X.shape[1], data_X.shape[2], 1)

	return data_X, classification


# this line will be executed first
#load data
trainX, trainy = load_full_dataset('train')
testX, testy = load_full_dataset('test')

print(trainX.shape), print(trainy.shape)
```

# Model Architecture Explanation

The model architecture is a convolutional neural network (CNN) designed for tasks such as binary classification. Below is a detailed explanation of each layer, its use case, activation function, and kernel configuration.

---

### 1. **Conv2D Layer (9 filters, kernel size (4, 1), ReLU activation)**
- **Purpose**: Extracts low-level features like edges, textures, and patterns from the input.
- **Kernel Size (4, 1)**: A 2D filter that scans 4 rows and 1 column at a time, useful for data with vertical structures or temporal dependencies (e.g., time series or spectrograms).
- **Number of Filters (9)**: Produces 9 unique feature maps.
- **ReLU Activation**: The Rectified Linear Unit outputs \( \text{max}(0, x) \), introducing non-linearity, preventing vanishing gradients, and improving training efficiency.

---

### 2. **Dropout Layer (rate 0.25)**
- **Purpose**: Prevents overfitting by randomly setting 25% of the layer's outputs to zero during training. This forces the model to generalize better by not overly relying on specific neurons.

---

### 3. **Conv2D Layer (3 filters, kernel size (1, 3), ReLU activation)**
- **Purpose**: Extracts higher-level features from the feature maps generated by the first Conv2D layer.
- **Kernel Size (1, 3)**: A narrow horizontal kernel scans 1 row and 3 columns at a time, capturing horizontal patterns or dependencies.
- **Number of Filters (3)**: Produces 3 feature maps for focused feature extraction.
- **ReLU Activation**: Adds non-linearity, ensuring effective gradient flow.

---

### 4. **Dropout Layer (rate 0.25)**
- **Purpose**: Prevents overfitting, similar to the previous Dropout layer, by dropping 25% of neurons during training.

---

### 5. **Flatten Layer**
- **Purpose**: Converts 2D feature maps into a 1D vector to prepare the data for fully connected (Dense) layers.
- **Reason**: Fully connected layers require input in a flat, vectorized form. This layer bridges the convolutional and dense layers.

---

### 6. **Dense Layer (30 units, ReLU activation)**
- **Purpose**: A fully connected layer that learns complex patterns and relationships from the extracted features.
- **Units (30)**: Specifies the number of neurons, allowing the model to learn multiple features.
- **ReLU Activation**: Ensures non-linear transformations, enabling the network to learn complex mappings.

---

### 7. **Dropout Layer (rate 0.2)**
- **Purpose**: Reduces overfitting by randomly setting 20% of the layers outputs to zero, encouraging generalization.

---

### 8. **Dense Layer (30 units, ReLU activation)**
- **Purpose**: Another fully connected layer for further refinement of feature representations.
- **Units (30)**: Same as the previous Dense layer, ensuring the model captures a richer feature set.
- **ReLU Activation**: Maintains non-linear transformations for effective learning.

---

### 9. **Dropout Layer (rate 0.2)**
- **Purpose**: Prevents overfitting again by randomly dropping 20% of the outputs during training.

---

### 10. **Dense Output Layer (Sigmoid Activation)**
- **Purpose**: Produces the final output of the model, representing the probability of the input belonging to a specific class.
- **Units (1)**: A single neuron is used as this is a binary classification task.
- **Sigmoid Activation**: Squashes the output into the range \([0, 1]\), representing probabilities. It is ideal for binary classification tasks.

---

## Summary of Layer Interactions and Use Cases
1. **Conv2D Layers**: Extract spatial and structural features from the input data.
2. **Dropout Layers**: Prevent overfitting and enhance generalization.
3. **Flatten Layer**: Bridges convolutional layers with dense layers by converting feature maps into a flat vector.
4. **Dense Layers**: Capture complex feature relationships and learn hierarchical representations.
5. **Activation Functions**:
   - **ReLU**: Efficient for intermediate layers, avoids saturation, and prevents vanishing gradients.
   - **Sigmoid**: Ideal for binary classification, outputs probabilities.
6. **Kernel Sizes**: Allow the model to learn patterns specific to vertical or horizontal dependencies in the data.

This architecture is compact yet versatile, suitable for tasks like image classification, time-series analysis, or binary decision-making problems.


```python

# fit and evaluate a model
def evaluate_model_2dconv(trainX, trainy, testX, testy):
	"""
	Evaluate a 2D Convolutional Neural Network model on the given training and testing data.

	This function builds, compiles, and trains a 2D CNN model using the provided training data.
	It then evaluates the model on the test data and returns the predictions along with their probabilities.

	Parameters:
	trainX (numpy.ndarray): Training data features.
	trainy (numpy.ndarray): Training data labels.
	testX (numpy.ndarray): Testing data features.
	testy (numpy.ndarray): Testing data labels.

	Returns:
	pandas.DataFrame: A DataFrame containing the predicted labels, their probabilities, and the true labels.

	Model Architecture:
	- Conv2D layer with 9 filters, kernel size (4,1), ReLU activation
	- Dropout layer with rate 0.25
	- Conv2D layer with 3 filters, kernel size (1,3), ReLU activation
	- Dropout layer with rate 0.25
	- Flatten layer
	- Dense layer with 30 units, ReLU activation
	- Dropout layer with rate 0.2
	- Dense layer with 30 units, ReLU activation
	- Dropout layer with rate 0.2
	- Dense output layer with sigmoid activation

	Training Configuration:
	- Loss function: binary_crossentropy
	- Optimizer: adam
	- Metrics: accuracy
	- Early stopping based on validation accuracy with patience of 50 epochs

	Example:
	>>> results = evaluate_model_2dconv(trainX, trainy, testX, testy)
	>>> print(results.head())
	"""
	verbose, epochs, batch_size = 1, 300, 1
	n_outputs = trainy.shape[1]
	model = Sequential()

	model.add(Conv2D(filters=9, kernel_size=(4,1), input_shape=trainX.shape[1:],activation='relu'))
	model.add(Dropout(0.25))
	model.add(Conv2D(filters=3, kernel_size=(1,3), activation='relu')) #(1,5)

	model.add(Dropout(0.25))

	model.add(Flatten())
	model.add(Dense(30, activation='relu'))
	model.add(Dropout(0.2))
	model.add(Dense(30, activation='relu'))
	model.add(Dropout(0.2))
	model.add(Dense(n_outputs, activation='sigmoid'))
	model.summary()
	model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

	es = EarlyStopping(monitor='val_accuracy',
				   mode='max',
				   patience=50,
				   restore_best_weights=True)

	history = model.fit(trainX, trainy, epochs=epochs, batch_size=batch_size, verbose=verbose,validation_split=.2,callbacks=[es])

	# evaluate model
	_, accuracy = model.evaluate(testX, testy, batch_size=batch_size, verbose=1)

	pred_label_ = model.predict(testX,batch_size=batch_size)

	pred_label = [1. if x >= 0.5 else 0. for x in pred_label_]
	results = pd.DataFrame({'Pred': pred_label,'Prob':pred_label_.reshape(-1), 'True': testy.reshape(-1)})

	if save_model:
		model.save('{}/my_model.keras'.format(model_path))

	return accuracy, history




```

```python
# summarize scores
def summarize_results(scores):
	print(scores)
	m, s = mean(scores), std(scores)
	print('Accuracy: %.3f%% (+/-%.3f)' % (m, s))

# run an experiment
def run_experiment(repeats=1):
	"""
	Run the experiment for a specified number of repeats.

	This function loads the training and testing datasets, evaluates the model using a 2D convolutional neural network,
	and summarizes the results. The experiment is repeated for the specified number of times.

	Parameters:
	repeats (int): The number of times to repeat the experiment. Default is 1.

	Returns:
	None
	"""
	#load data
	trainX, trainy = load_full_dataset('train')
	testX, testy = load_full_dataset('test')


	# repeat experiment
	scores = list()
	for r in range(repeats):
		score,history = evaluate_model_2dconv(trainX, trainy, testX, testy)

		score = score * 100.0
		print('>#%d: %.3f' % (r+1, score))
		scores.append(score)

	summarize_results(scores)


run_experiment()
```
